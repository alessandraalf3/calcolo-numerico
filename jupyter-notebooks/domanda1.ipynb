Dato un problema test di dimensioni variabili 
$$
Ax=b
$$

la cui soluzione esatta è il vettore $x=(1,...,1)^T$ e $b$ è il termine noto, ottenuto come $Ax$ calcolare:
- il numero di condizionamento della matrice $A$
- la soluzione $x$ del sistema lineare $Ax=b$ mediante la fattorizzazione $LU$ con pivoting.

Considerare come matrice $A$ la matrice quadrata, di numeri casuali generati mediante la funzione ``randn`` di Numpy, con $n$ variabile tra 10 e 1000. 

```python
import numpy as np
import scipy
import scipy.linalg

n = np.random.randint(10, 1001)

A = np.random.randn(n, n)
x = np.ones((n, 1))
b = A @ x


print('x.shape: ', x.shape, '\n' )

print('b.shape: ', b.shape, '\n' )

print('A.shape: ', A.shape, '\n' )

print('K(A)=', np.linalg.cond(A), '\n')
```

```python
lu, piv = scipy.linalg.lu_factor(A)

print('lu',lu,'\n')
print('piv',piv,'\n')


my_x = scipy.linalg.lu_solve((lu, piv), b)

print('my_x = \n', my_x)
```
